```python
import pandas as pd
from random import *

#df = pd.read_csv("kaoqin.csv")
df = pd.read_excel("2021.2.xls")
df_names = df["Name"]
name_list = df_names.unique().tolist()
```

```python
month = df["DateTime"][0].month
year = str(2021)
checkin_times = [(6, 49), (11, 30), (12, 20), (17, 0)]
outputDataframe = pd.DataFrame()
```

```python
df
```

```python
for staff in name_list:
    print(staff)
    for day in range(1, 29):
        for (hour, minute) in checkin_times:
            #print(f"{staff} {month}-{day}/{year} {hour}:{minute}\n")
            mode = "指纹"
            if randint(1, 32) == 16:
                mode = "刷脸"
            outputDataframe = outputDataframe.append({"Name": staff, "Mode": mode, "DateTime": pd.to_datetime(f"{year}/{month}/{day} {hour}:{str(minute+randint(1,10)).zfill(2)}:{randint(0,59)}", format="%Y/%m/%d %H:%M:%S")}, ignore_index=True)
print(outputDataframe)            
```

```python
outputDataframe.to_excel("new-2021.2.xlsx", index=False)
```
