```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import json
```

```python
db = pd.read_csv('tmdb_5000_movies.csv')
```

```python
db.sample()
```

```python
selected_features = ['genres']
db2 = pd.DataFrame()
db2[selected_features] = db[selected_features]
db2['id'] = db['id']
```

```python
db2.sample()
```

```python
db2['genres'] = db2['genres'].apply(func=lambda x: ' '.join(map(lambda y: y['name'], json.loads(x))))
```

```python
db2.sample()
```

```python
tfidf = TfidfVectorizer()
vec = tfidf.fit_transform(db2['genres'])
vec.todense()
```

```python
similarities = cosine_similarity(vec)
```

```python
similarities
```

```python
origin_movie_name = input("Enter the movie's name.")
```

```python
db.sample()
```

```python
origin_id = db.loc[db['original_title'] == origin_movie_name].iloc[0]['id']
```

```python
origin_id
```

```python
indexInDb2 = db2.index[db2['id'] == origin_id]
```

```python
indexInDb2
```

```python
indexInDb2 = indexInDb2[0]
```

```python
similarities_list = []
for index, item in enumerate(similarities[indexInDb2]):
    similarities_list.append((item, index))
similarities_list.sort(key=lambda x: x[0], reverse=True)
```

```python
similarities_list
```

```python
for _, i in similarities_list[:10]:
    print(db[db['id'] == i]['original_title'])
```
