```python
import pandas as pd
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
db = pd.read_csv('target_db.csv', index_col='title')
db.drop(axis=1, columns=[db.columns[0]], inplace=True)
db = db.astype(str)
db
```

```python
db[pd.isna(db['genres'])] = 'common'
```

```python
db.columns
```

```python
db2 = pd.DataFrame()
db2['total'] = db['genres'].map(lambda x: " ")
for i in db.columns:
    db2['total'] = db2['total'] + ' ' + db[i]
db2
db2.to_csv('temp.csv')
db = db2.copy()
```

```python
from itertools import combinations
tf = TfidfVectorizer()
result_matrix = tf.fit_transform(db['total'])
result_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities = cosine_similarity(result_matrix)
similarities
```

```python
sim2 = pd.DataFrame(similarities)
sim2.columns = db.index
sim2.index = db.index
# sim2
```

```python
pd.DataFrame(result_matrix.todense(), columns=tf.get_feature_names(), index=db.index).sample(10, axis=0).sample(10, axis=1)
```
